## Metric used for this file is arrest rate.

### Introduction
Since we are finding the best neighborhood in Pittsburgh, the first thing I was thinking about that is the most important thing is crime rate. I set arrest rate as my submetric and searched the dataset of arrest counts at each neighborhood from 2015-2022. To calculate the arrest rate, I divided the arrest counts by area of neighborhood in the final notebook to get the arrest rate at each unit area preventing large area has larger arrest count issue.<br>

### My metric
My submetric is the arrest rate at each unit area for each neighborhood. I used the dataset called Pittsburgh Police Arrest Data that having Pittsburgh arrest counts for all neighborhood between 2015-2022.<br>

### The best neighborhood
After apply my metric, the best neighborhood is Mt. Oliver Neighborhood	in Pittsburgh. It only has 2 arrest count between these years therefore based on my metric, it is the safest place to live in. I also came up with a visualization bar graph that shows the lowest arrest number is 2 at Mt. Oliver Neighborhood.<br>



```python
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
```

```python
# read the downloaded csv file
arrest_df = pd.read_csv("Pittsburgh_Police_Arrest_Data.csv", sep=',', engine='python')

arrest_df
```

```python
# extract the INCIDENTNEIGHBORHOOD column only and count arrests counts by adding up all the arrest incidents counts for each neighborhood, sort them in descending order
# make a dataframe
value_counts = arrest_df['INCIDENTNEIGHBORHOOD'].value_counts(dropna=True, sort=True)
df = pd.DataFrame(value_counts).reset_index()
df.columns = ['neighborhood', 'incidents_counts']
df
```

```python
# save to csv file for future use in final notebook
df.to_csv('arrest_data.csv')
```

### Conclusion
By collecting, merging and anaylizing data, the neighborhood with the lowest arrest count is Mt. Oliver Neighborhood. Therefore based on my metric, Mt. Oliver Neighborhood is the best neighborhood. ### Visualized bar diagram

```python
# Initializing a large figure
arrest_plt = plt.figure(figsize=(40, 24))
plt.ticklabel_format(style='plain')
plt.bar(range(len(df['neighborhood'])),df['incidents_counts'], align = 'center')

# Formatting graph
plt.xticks(range(len(df['neighborhood'])),df['neighborhood'], 
           rotation = 45, fontsize = 8)
plt.xlabel('Pittsburgh Neighborhood Location')
plt.ylabel('Arrests Counts')
plt.title('Pittsburgh Neighborhood Arrests 2016-2022')

# save graph
plt.savefig("bar_plot.png")
plt.show()
```
