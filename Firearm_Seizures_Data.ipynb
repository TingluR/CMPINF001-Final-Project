# My Metric is Firearn Seizures dataset

I determine whether a neighborhood is the best neighborhood in Pittsburgh from 2015-2022 by the total count of firearm seizures, additionaly the farthest away from the present is the best.

```python
#load the CSV file into the DF
import pandas as pd

df = pd.read_csv('Firearm_Seizures_Data.csv', index_col=0)
df
```

```python
#The second step was to combine the same neighborhood and add together them total_count from 2015 to 2022  to generate new data
total = df.groupby('neighborhood').agg(Firearm_Seizures = ('total_count', 'sum'))

total.sort_values("Firearm_Seizures", ascending=False).to_csv('TheFirearmS.csv')

total.sort_values("Firearm_Seizures", ascending=False)
```

```python
#Since these four neighborhood fiream seize count are same, so I want go ferther to know which one is the best. I sort location by year columns with descending order.
#Compared to other dates. the timelength of Summer Hill is the farthest away from the present.
filtered_df = df.loc[(df['neighborhood'] == 'Summer Hill')]
filtered_df.sort_values("year", ascending=False)
```

```python
filtered_df1 = df.loc[(df['neighborhood'] == 'Swisshelm Park')]
filtered_df1.sort_values("year", ascending=False)
```

```python
filtered_df2 = df.loc[(df['neighborhood'] == 'Mt. Oliver')]
filtered_df2.sort_values("year", ascending=False)
```

```python
filtered_df3 = df.loc[(df['neighborhood'] == 'Regent Square')]
filtered_df3.sort_values("year", ascending=False)
```

```python
import matplotlib.pyplot as plot

total["neighborhood"] = total.index #index

total.plot.bar(x="neighborhood", y="Firearm_Seizures", rot=90,figsize=(20,10), title="Fire Inciden 2015-2022")

```
