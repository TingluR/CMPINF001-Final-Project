# My Metric is Firearn Seizures dataset

```python
import pandas as pd

df = pd.read_csv('Firearm_Seizures_Data.csv', index_col=0)
df
```

```python
total = df.groupby('neighborhood').agg(Total = ('total_count', 'sum'))

total.sort_values("Total", ascending=False).to_csv('TheFirearmS.csv')

total.sort_values("Total", ascending=False)#.head(5)#True Faluse
#total.sort_values("Total", ascending=False).head(5)
```

```python
import matplotlib.pyplot as plot

total["neighborhood"] = total.index #index

total.plot.bar(x="neighborhood", y="Total", rot=90,figsize=(20,10), title="Fire Inciden 2015-2022")

```
