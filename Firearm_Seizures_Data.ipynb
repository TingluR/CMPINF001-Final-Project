```python
import pandas as pd

df = pd.read_csv('Firearm_Seizures_Data.csv', index_col=0)
df
```

```python
total = df.groupby('neighborhood').agg(Total = ('total_count', 'sum'))
total.sort_values("Total", ascending=True)#.head(5)#True Faluse
#total.sort_values("Total", ascending=False).head(5)
```

```python
import matplotlib.pyplot as plot

total["hood"] = total.index #index

total.plot.bar(x="hood", y="Total", rot=90,figsize=(20,10), title="Number of tourist visits - Year 2018");

```
